```python
import pandas as pd

df = pd.read_csv('pays.csv', sep=',', index_col=0)
df.head()
```

```python
# b) variables numériques stockées dans un tableau num_data
num_data = df.select_dtypes(include=['int64', 'float64'])
print(num_data)
```

```python
# d) moyenne de chaque variable dans num_data stockée dans une seule colonne.
stats = pd.DataFrame(num_data.mean(), columns = ['mean'])
print(stats)
stats.round(2)
```

```python
# e) comparaison entre moyenne taux d'urbanisation et moyenne taux de natalité
## Les deux sont homogènes
# Ces deux moyennes sont à peu prs égales, on peut prendre l'hypothèse que ces deux variables présente une homogenity dans l'intervale
# Mais ce n'est pas suffisant pour cette determination, pour cette raison il faut calculer la médiane et les quantiles.
## La mediane represente 50% de mon échantillon
```

```python
# f)
stats['mediane'] = num_data.median()
stats
```

```python
# g) Interprétation des valeurs
# On peut conclure pour le taux d'urbanisation 50% de notre échantillon présentent un taux inferieur de 29.8
# Aussi bien 5% de notre échantillon representent un taux de natali inférieur/supérieur à 31.8
# La différence entre la myenne et la méd n'est pas énorme et alors on peut toutjours conclure l'hypothèse que ces deux variables
# presentent une homogenity des valeurs dans les intervalles comparable.
```

```python
# h) variable mean_med_diff correspondant à la valeur absolue de la différence entre la moyenne et la médiane
stats['mean_med_dif'] = abs(stats['mean'] - stats['mediane'])
```

```python
# i) quantiles d'objectif
num_data.quantile(q = [0.25, 0.5, 0.75])
```

```python
# j)
stats[['q1', 'q2', 'q3']] = num_data.quantile(q = [0.25, 0.5, 0.75]).transpose()
```

```python
# l) comparaison
# Si on compare les valeurs des quantiles et alors q1 avec q2 et q2 avec q3, on se rend compte que l'écart entre ces valeurs n'est pas énorme
# et ils sont de même grandeur. Même en comparant le troisième quantile à la moyenne, on est toujours dans la même zone de valeurs
# cela confirme notre hypothèse de base que les points sont disperser dans des intervales comparables
# et la dispersion des points est homogène.
```

```python
# m)
stats['min'] = num_data.min()
stats['max'] = num_data.max()
stats['min_max_diff'] = stats['max'] - stats['min'] 
```
